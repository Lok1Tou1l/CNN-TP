# Imports

```python
import torch
import torch.nn as nn
import torch.optim as optim
import os
device = torch.device('cpu')
```

# Parameters

```python
DATA_DIR = 'data/images/'
BATCH_SIZE = 32
NUM_CLASSES = len(os.listdir(DATA_DIR))
NUM_EPOCHS = 10
LEARNING_RATE = 0.001
```

# Data preprocessing 

```python
from src import data_preprocessing as dp
train_loader , test_loader = dp.preprocess_data(DATA_DIR,BATCH_SIZE,0.2)
```

# Model definition

```python
from src import model as net
model = net.SimpleCNN(NUM_CLASSES).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = LEARNING_RATE   )
```

# Training Loop

```python
for epoch in range(NUM_EPOCHS) :
    model.train()
    running_loss = 0.0
    for i, (inputs, labels) in enumerate(train_loader):
        inputs , labels = inputs.to(device) ,labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs , labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    epoch_loss = running_loss / len(train_loader)
    print(f'Epoch [{epoch + 1}/{NUM_EPOCHS}], Loss: {epoch_loss:.4f}')

print('Training Finished!')
```

# Save the trained model in ..\models

```python
model_save_path = os.path.join('models', 'trained_model.pth')
torch.save(model.state_dict(), model_save_path)
print(f'Trained model saved to {model_save_path}')
```
